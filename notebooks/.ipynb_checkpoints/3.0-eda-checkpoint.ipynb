### Describe the correlation between features in the dataset (0.75 points)

Considering the available features within the training dataset, we want to identify and analyse the relationships between them, and then determine which features within the dataset significantly contribute to our solution goal. By using calculation and visualisation, Python allows us to create a correlation matrix, which is a table that represents the correlation coefficients between different variables.

Now let's make a correlation matrix, then visualize it, and describe what insights you observed?

```python
# PRE-PROCESSING IF NEEDED
# YOUR CODE HERE (OPTION) 
df_copy = cleaned_df.copy()
df_copy['Year'] = df_copy['Year'].astype('category').cat.codes
df_copy['Country'] = df_copy['Country'].astype('category').cat.codes
```

## Making your own question 01? (1.5 points)

**YOUR ANSWER HERE**

**Question**: How does the gender ratio affect the economy?\
**Answering this question will** help us to see the economic benefits of gender equality.\
**How we answer this question**: we will consider the average gender ratio for countries across all the year, then get the average GDP in USD of countries to compare by using gender ratio.

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 
pop_gdp_df = pd.DataFrame()
pop_gdp_df['Country'] = cleaned_df['Country']
pop_gdp_df['Gender Ratio'] = [cleaned_df['Female Population'][i] / cleaned_df['Male Population'][i]
                              for i in range(len(cleaned_df['Male Population']))]
pop_gdp_df['GDP in USD'] = cleaned_df['GDP in USD']

avg_pop_gdp_df = pop_gdp_df[['Gender Ratio', 'GDP in USD', 'Country']].groupby('Country').apply(lambda df_: df_.mean())
avg_pop_gdp_df
```

### Analyzing to get the answer

```python
# YOUR CODE HERE
# plotting all countries might have a problem about X-ray, so I plot the top 20 countries with the highest GDP
top20_gdp_df = avg_pop_gdp_df.nlargest(20, 'GDP in USD')
countries = top20_gdp_df.index.tolist()

gender_ratio = top20_gdp_df['Gender Ratio']
gdp = top20_gdp_df['GDP in USD']

palette = sns.color_palette("RdBu", len(gdp))

plt.figure(figsize=[25, 15])
plt.bar(countries, gender_ratio, color = palette, width = 0.8)

plt.ylabel('Gender Ratio')
plt.ylim(0, gender_ratio.max() + 0.5)
plt.grid(True)
plt.title('How the gender ratio affect to the economy')
plt.show()
```

**MY ANSWER**
According to the plot, we can see that most high-GDP countries have a gender ratio of almost 1.\
This observation can be explained for some reasons:
- The gender equality promotes the inclusivity and diversity. The work can be operated more productively.
- It also helps to reduce the economic burden for a family that has only one economic pillar.
- Additionally, gender equality can expand consumer markets and increase the talent pool.

## Making your own question 02? (1.5 points)

**YOUR ANSWER HERE**
**Question**: What is the level of renewable energy use in high-GDP countries?\
**Answering this question will** helps us to see the level of energy use conversion of the high-GDP countries.\
**How we answer this question**: We will get the top 5 countries with the highest average `Renewable Energy Consumption (%)`.\
*Note*: I will remove 2 years 2021 and 2022 because rows which contain these years in the original dataset have NaN values on `Renewable Energy Consumption (%)` columns. 

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 
# REC = Renewable Energy Consumption
REC = cleaned_df[['Renewable Energy Consumption (%)', 'Year', 'Country']].set_index('Year')
REC = REC.drop([2021, 2022], axis = 'index')

avg_REC = REC[['Renewable Energy Consumption (%)', 'Country']].groupby('Country').apply(lambda df_: df_.mean()) 
top10_avg_REC = avg_REC.nlargest(10, 'Renewable Energy Consumption (%)')
top10_countries = top10_avg_REC.index.tolist()

top10_avg_REC = cleaned_df.loc[cleaned_df['Country'].isin(top10_countries)][['Renewable Energy Consumption (%)', 'Year', 'Country']]
top10_avg_REC = top10_avg_REC.set_index('Year').drop([2021, 2022], axis = 'index')
top10_avg_REC = top10_avg_REC.reset_index()
top10_avg_REC
```

### Analyzing to get the answer

```python
# YOUR CODE HERE
plt.figure(figsize=[20, 10])

for i in range(10):
    country_df = top10_avg_REC.loc[top10_avg_REC['Country'] == top10_countries[i]]
    plt.xticks(country_df['Year'], country_df['Year'])
    plt.plot(country_df['Year'], country_df['Renewable Energy Consumption (%)'], '-o', label = top10_countries[i])
    
plt.xlabel('Year')
plt.ylabel('Renewable Energy Consumption (%)')
plt.title('Renewable Energy Consumption by Years')

plt.grid(True)
plt.legend()
plt.show()
```

**MY ANSWER**\
According to the plot, we can see the conversion of using renewable energy of each country is increasing, except Albania.\
Specially, the country with the strongest conversion is Iceland, a country with a harsh climate and the abundant natural resources of renewable energy. That's what makes Iceland one of the cleanest countries in the world.\
In conclusion, most European countries are switching to using more renewable energy, the reason is probably to save other gas resources and protect the environment.
