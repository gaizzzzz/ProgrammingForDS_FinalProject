## Question 1: Does the runtime of a movie affect audience ratings?

### Step 0: Purpose
- Help us to see the audience's preferences in terms of time spent watching movies, thereby helping filmmakers adjust the movie length accordingly.

### Step 1: Analyzing
- We will consider the `Rating (Out of 10)` column, divide it into 4 bins to see the distribution of user ratings. To give observation, we will visualize the distribution with scatter plot.

```python
rating_on_runtimes = cleaned_df[['Runtime (Minutes)', 'Rating (Out of 10)']]
```

```python
min_rating = rating_on_runtimes['Rating (Out of 10)'].min()
max_rating = rating_on_runtimes['Rating (Out of 10)'].max()
bins = [(min_rating, 5.0), (5.1, 7.0), (7.1, 8.5), (8.6, max_rating)]

runtimes_per_bin = dict.fromkeys(bins)

for bin in bins:
    runtimes = []
    ratings = []
    for i, row in rating_on_runtimes.iterrows():    
        if bin[0] <= row['Rating (Out of 10)'] <= bin[1]:
            ratings.append(row['Rating (Out of 10)'])
            runtimes.append(row['Runtime (Minutes)'])
    runtimes_per_bin[bin] = (ratings, runtimes)
```

### Step 2: Visualization

```python
plt.figure(figsize = (15, 10))

for key, value in runtimes_per_bin.items():
    plt.scatter(x = value[0], y = value[1], alpha = 0.5, label = key)

plt.title('Relationship of Movie Runtime vs. Rating')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Rating (Out of 10)')
plt.xticks(range(0, 11))
plt.legend(loc = 'upper left')
plt.grid(True)
plt.show()
```

```python
corr = cleaned_df['Runtime (Minutes)'].corr(cleaned_df['Rating (Out of 10)'])
print(f'Correlation Coefficient: {corr}')
```

### Step 3: Give obervation
- Movies with a duration of 100 to 200 minutes make up the majority and are concentrated in the rating range of 7 to 8.5.
- Movies with shorter durations have lower scores.
- Movies with high durations (greater than 200 minutes) account of very little, and most of them has the high ratings. I think they might be TV series.
- According to the correlation coefficient of `0.367314173636608` indicates a positive correlation between the two variables being compared, but the correlation is not very strong. The correlation coefficient suggests a moderate positive relationship, meaning that as the `Runtime` variable increases, the `Rating` variable tends to increase, but the increasement is not dramatic. Therefore, the relationship between two variables depends on many other factors, such as the popularity of the cast, director, filmmaker, etc.
- In conclusion, the movie rating tends to increase gently when the movie runtime increases. Moreover, I think the production should adjust the movie duration of 100 to 200 minutes to meet the audience's habits.d.

## Question 2: What movie genres work well together?

### Step 0: Purpose
- Help production knows which movie genres can be easily combined, thereby increasing the company's profits based on popular genres that audiences often choose to watch.

### Step 1: Analyzing
- We will consider the `Name` and `Genre` columns. We create a matrix with row and column labels are `genres`. We then count the number of common occurrences of each pair of categories and save them in matrix. Finally, we visualize by drawing a heatmap with the lower half of the matrix.

```python
# preprocessing
fixed_df = cleaned_df[['Title', 'Main Genres']].reset_index(drop = True)
fixed_df['Main Genres'] = fixed_df['Main Genres'].str.split(',')
fixed_df = fixed_df.explode('Main Genres')
fixed_df = fixed_df.query("`Main Genres` == `Main Genres`")
```

```python
combination_df = pd.DataFrame(index = fixed_df['Main Genres'].unique(), columns = fixed_df['Main Genres'].unique())

# assign the count of combination
for i in range(len(fixed_df['Main Genres'].unique()) - 1):
    for j in range(i + 1, len(fixed_df['Main Genres'].unique())):
        temp = fixed_df[(fixed_df['Main Genres'] == fixed_df['Main Genres'].unique()[i]) | 
                        (fixed_df['Main Genres'] == fixed_df['Main Genres'].unique()[j])].index
        combination_df.at[fixed_df['Main Genres'].unique()[i], fixed_df['Main Genres'].unique()[j]] = temp.shape[0] - temp.nunique() if temp.shape[0] - temp.nunique() else 0
combination_df = combination_df.T.fillna(0)
combination_df
```

### Step 2: Visualization

```python
plt.figure(figsize = (20, 15))
mask = np.triu(np.ones_like(combination_df), k = 1)
sns.heatmap(combination_df, annot = True, fmt = 'd', cmap = 'Greens', mask = mask, vmin = 0, vmax = 1000)
plt.show()
```

### Step 3: Give Observation
- According to the heatmap, we can see some pairs of genres have combined regularly:
    - `Action` and `Adventure`.
    - `Drama` and `Romance`.
    - `Drama` and `Comedy`.
    - `Drama` and `Crime`.
- That's understandable because their nature is the same. One side is a movie genre that brings dramatic suspense, the other side brings laughter or a feeling of relaxation.
- In conclusion, we recommend that filmmakers should make genre combination as below:
    - `Action` and `Adventure`.
    - `Drama`, `Romance` and `Comedy`.
    - `Drama` and `Crime`.
